```python
df1 = pd.DataFrame({
    'Item':['A','B','C'],
    'Price' : [30,40,80],
})
df2 = pd.DataFrame({
    'Country':['X','Y','Z'],
    'Available' : [True,True,False],
})


pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df3 = pd.DataFrame({
    'Item':['B','C','D'],
    'Country':['X','Y','Z'],
})

pd.merge(df1,df3)   # default : inner
```

```python
pd.merge(df1,df3,how='outer')
```

```python
pd.merge(df1,df3,how='left')
```

```python
pd.merge(df1,df3,how='right')
```

```python
pd.merge(df1,df3,on='Item',how='outer')
```
